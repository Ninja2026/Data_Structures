## Question 2

### Explain the difference between mutable and immutable data types with examples.

**Mutable Data Types**

Definition: A mutable object can be modified after it is created. We can change its content, add new elements, or remove existing ones without changing its identity (memory address).

Key Characteristic: Changes happen "in-place."

Examples: Lists, Dictionaries, Sets.

```python
my_list = [1, 2, 3]
my_list[0] = 99  
print(my_list)
```

**Immutable Data Types**

Definition: An immutable object cannot be changed once it is created. If we try to modify it, the program actually creates a brand-new object in memory with the new value.

Key Characteristic: Any "change" results in a new object with a new identity.

Examples: Integers, Floats, Strings, Tuples.

```python
my_tuple = (1, 2, 3)

try:
    # This will trigger a TypeError:
    my_tuple[0] = "99"
except TypeError as e:
    print(f"Tuple Error:   {e}")
```

## Question 17

### Describe the time complexity of accessing elements in a dictionary?

The time complexity of accessing an element in a dictionary is $O(1)$, also known as Constant Time.This means that whether our dictionary has 10 items or 10 million items, the time it takes to retrieve a value remains virtually the same. This is the "superpower" of dictionaries that makes them the go-to structure for high-performance data retrieval.

## Question 19

### Why are dictionaries considered unordered, and how does that affect data retrieval?

Dictionaries are considered unordered because they are implemented as hash tables designed for fast, key-based,$O(1)$ lookup speed rather than maintaining insertion order

It's affects on data retrieval are :-

- **Instant Access ($O(1)$):** In a list (ordered), the computer might have to check every item to find what you're looking for. In a dictionary, it simply hashes your key and goes straight to the correct "bucket" in memory.

- **No Indexing:** You cannot retrieve the "first" or "last" item using a position. You must know the Key to get the Value.

- **Predictable Speed:** Whether the dictionary has 10 items or 10 million, finding a specific key takes roughly the same amount of time.
